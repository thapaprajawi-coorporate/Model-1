```python
#this is for calculating the balloon payment
P = float(input("The total amount of money that you loaned out: "))
R = float(input("The interest rate on said loan in decimal:"))
T = int(input("The term of the loan in months:"))
CFDA = []
Discounted_CFDA = []
for i in range(T):
    cash_flow = float(input( f"Enter the Cash FLow Available for Debt Services for month {i+1}:"))
    discounted_cash_flow = ( cash_flow / ( 1 + R ) ** ( i + 1 ))
    CFDA.append(cash_flow)
    Discounted_CFDA.append(discounted_cash_flow)
    
print("Your CFDAS for the term of the loan is:",CFDA)
print("Your DCFDAS for the term of the loan is:",Discounted_CFDA)

mr = R / 12                                            
i = 0
interest = []
while i < T:
    interest_current_month = P * mr
    interest.append(interest_current_month)
    if CFDA[i] - interest_current_month >= P:
        principal_paid_current = P
        ending_balance = 0
        print(f"Loanhas been paid off early in month {i+1}")
    else:
        principal_paid_current = CFDA[i] - interest_current_month
        ending_balance = P - principal_paid_current
    P = ending_balance
    print(f"Principal paid for month {i + 1}: {principal_paid_current}")
    i = i + 1
balloon = ending_balance
print(" Final balloon payment:", balloon) 

#calculating the ICR: Interest Coverage Ratio

total_interest = sum(interest)
earning_before_tax_interest = sum(CFDA) 

ICR = earning_before_tax_interest/total_interest

print(f"The Interest Coverage ratio is:{ICR}")

if ICR < 3:
    print(f"Warning: The proposed loan puts the business's ICR is at: {ICR}. The business may struggle to refinance the balloon payment under current cashflow conditions")
else:
    print(f"Based on current cash flows, the business maintains a healthy ICR of :{ICR}. This indicates a strong likelihood of qualifying for refinancing when the balloon payment is due")

# calculating simple pay back time

total = 0
for i in range(T):
    total = total + CFDA [ i ]
    if total >= P:
        print(f"The simple payback period is:{i+1}")
        paid_off = True 
        break 
        
if not paid_off:
    print("The loan principal is not fully recovered within the term")

# calculating discounted pay back time 

discounted_total = 0

for i in range(T):
    discounted_total = discounted_total + Discounted_CFDA[i]
    if discounted_total >= P:
        print(f"The discounted payback period is:{i+1}")
        paid_off = True 
        break 
        
if not paid_off:
    print("The loan principal is not fully recovered within the term")

    
print ( f"The business will be able to payback in:{i-1} years" )

#calculating net present value 




```
