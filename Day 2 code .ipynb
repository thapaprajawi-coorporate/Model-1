```python
P = float(input("The total amount of money that you loaned out: "))
R = float(input("The interest rate on said loan in decimal:"))
T = int(input("The term of the loan in months:"))
CFDA = []
for i in range(T):
    cash_flow = float(input( f"Enter the Cash FLow Available for Debt Services for month {i+1}:"))
    CFDA.append(cash_flow)
print("Your CFADS for the term of the loan is:",CFDA)

mr = R / 12                                            
i = 0
while i < T:
    interest_current_month = P * mr
    principal_paid_current = CFDA[i] - interest_current_month
    ending_balance = P - principal_paid_current
    P = ending_balance
    print(f"Principal paid for month {i + 1}: {principal_paid_current}")
    i = i + 1
balloon = ending_balance
print(" Final balloon payment:", balloon) 

gross_revenue = float(input("Annual total sales: "))
operating_expenses = float(input("Total cost to run the business : excluding taxes/interests:"))
existing_debts = float(input("Current aannual loans or lease payments the business already has:"))
new_loan_payment = sum(CFDA) / 12 # to make it an annual average to make the math correct 

Net_Operating_Income = gross_revenue - operating_expenses
Proposed_Debt_Services = existing_debts + new_loan_payment

DSCR = Net_Operating_Income / Proposed_Debt_Services

if DSCR < 1.25:
    print(f"Warning: The proposed loan puts the business's DSCR at: {DSCR}. The business may struggle to refinance the balloon payment under current cashflow conditions")
else:
    print(f"Based on current cash flows, the business maintains a healthy DSCR of :{DSCR}. This indicates a strong likelihood of qualifying for refinancing when the balloon payment is due")







```
